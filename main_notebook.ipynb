# Using cats api

# This API contains cats info so it could be used for an app to daily spam a cats fact to the user

```python
import requests
import json
import pandas as pd
```

```python


cats = requests.get("https://cat-fact.herokuapp.com/facts")
print("Cats:", cats.status_code)
```

```python
print(cats.text)
```

```python


results = cats.json()
results
```

```python


data = pd.DataFrame(results)
data
```

```python
data.dtypes
```

```python
pd.set_option('display.max_colwidth', None)
data['text']
```

# Dota players by rank api

# Dota is a game. A Dota 2 player rank API can provide valuable information about the rankings and statistics of Dota  players.
# For example: Build a leaderboard system that ranks players based on their MMR or other performance metrics. This can encourage healthy competition within the Dota 2 community.

```python


dota_matches = requests.get("https://api.opendota.com/api/playersByRank")
print("Dota:::", dota_matches.status_code)
```

```python


results = dota_matches.json()
results
```

```python


data = pd.DataFrame(results)
data
```

```python
data.dtypes
```
